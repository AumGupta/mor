```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, chisquare, chi2

# Data
x = np.array([0, 1, 2, 3, 4])
o = np.array([32, 178, 290, 236, 64])

N = 800  # Total number of observations
n = 4    # Number of trials

# Binomial probabilities (expected probabilities for each outcome)
Pi = binom.pmf(x, n, 0.5)

# Expected frequencies (Pi * N)
E = Pi * N

# Perform chi-square test
chi_stat, p_value = chisquare(o, E)

# Critical chi-square value (for 4 degrees of freedom and 95% confidence)
chi_tab = chi2.ppf(0.95, 4)

# Print Results
print(f"Chi-Square Statistic: {chi_stat}")
print(f"P-Value: {p_value}")
print(f"Critical Value: {chi_tab}")

# Compare the test statistic to the critical value
if chi_stat > chi_tab:
    print("Reject the null hypothesis: Observed frequencies do not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed frequencies fit the expected distribution.")

# Plotting the observed vs expected frequencies

plt.figure(figsize=(10, 6))

# Bar plot for observed frequencies
plt.bar(x - 0.2, o, width=0.4, label='Observed', color='blue', align='center')

# Bar plot for expected frequencies
plt.bar(x + 0.2, E, width=0.4, label='Expected', color='orange', align='center')

# Adding labels and title
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.legend()

plt.show()

# Plotting the Chi-Square Distribution and the test statistic
x_vals = np.linspace(0, 15, 500)
y_vals = chi2.pdf(x_vals, 4)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Chi-Square Distribution (df=4)", color='green')

# Marking the test statistic and critical value
plt.axvline(chi_stat, color='red', linestyle='--', label=f"Chi-Square Statistic: {chi_stat:.2f}")
plt.axvline(chi_tab, color='purple', linestyle='--', label=f"Critical Value: {chi_tab:.2f}")

plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution and Test Statistic')
plt.legend()

plt.show()
```
